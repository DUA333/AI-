```python
dta=data.dropna(axis=0, how="any")
dta
```

```python
dta.isnull().sum()
```

```python
dta.info()
```

```python
gender = pd.get_dummies(dta[['gender']], drop_first= True)
gender.head()
```

```python
ever_married = pd.get_dummies(dta[['ever_married']], drop_first= True)
ever_married.head()
```

```python
 work = pd.get_dummies(dta[['work_type']], drop_first= True)
 work.head()
```

```python
residence = pd.get_dummies(dta[['Residence_type']], drop_first= True)
residence.head()
```

```python
smoking = pd.get_dummies(dta[['smoking_status']], drop_first= True)
smoking.head()
```

```python
dta.head()
```

```python
dat = pd.concat([dta, gender, ever_married, work, residence, smoking], axis = 1)
dat.head()
```

```python
dat = dat.drop(columns=categorical_features, axis=1)
dat.head()
```

```python

dat = dat.drop(columns="id" , axis=1)
dat.head()
```

```python
dat.columns
```

```python
dat.describe()
```

```python
corr = dat.corr()['stroke'].sort_values(ascending=False).to_frame()
plt.figure(figsize=(2,8))
sns.heatmap(corr, cmap='Blues', cbar=False, annot=True)
plt.show()
```

```python
x_cols = dat.drop("stroke", axis=1).columns
X = dat[x_cols]
y = dat['stroke']
```

```python
y.head()
```

```python
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X= scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=69)
```

```python
x_train
```

```python
from sklearn.tree import DecisionTreeClassifier
decision = DecisionTreeClassifier(random_state=1)
decision.fit(x_train, y_train)
```

```python
decision.score(x_test, y_test)
```

```python
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression()
lg.fit(x_train,y_train)
```

```python
lg.score(x_test,y_test)
```
